```python
# Import Dependencies
import requests
from selenium import webdriver
import time
import pandas as pd
import os
import getpass 
from sqlalchemy import create_engine
```

```python
# Request file path from user to save reports to
file_path = input("Please enter this file's file path: ")

# User Prompt
print('   Your reports will be downloaded here\n\n\
Time to sign in to data.world!')

# Request Username
username = input('     Username: ')

# Request Password 
try: 
    password = getpass.getpass('     Password: ') 
except Exception as error: 
    print('ERROR', error) 
```

```python
# URL to login
login_url = 'https://data.world/login'

# List of report URLs
reports_url = ['https://data.world/san-diego/ripa-police-stop-data-race-of-persons-stopped-datasd/\
workspace/file?filename=ripa-police-stop-data-race-of-persons-stopped-2.csv',
        
        'https://data.world/san-diego/ripa-police-stop-data-gender-of-persons-stopped-datasd/\
workspace/file?filename=ripa-police-stop-data-gender-of-persons-stopped-1.csv',
        
        'https://data.world/san-diego/ripa-police-stop-data-reason-for-stop-datasd/\
workspace/file?filename=ripa-police-stop-data-reason-for-stop-1.csv',
        
        'https://data.world/san-diego/ripa-police-stop-data-result-of-stop-datasd/\
workspace/file?filename=ripa-police-stop-data-result-of-stop-1.csv',
        
        'https://data.world/san-diego/ripa-police-stop-data-actions-taken-datasd/\
workspace/file?filename=ripa-police-stop-data-actions-taken-1.csv']
```

```python
# Create Chrome Options
options = webdriver.ChromeOptions()

# Create directory preferences and ignore 
preferences = {'download.default_directory':file_path,\
               'safebrowsing.enabled': 'false'}

# Set webdriver to save in preferences
options.add_experimental_option('prefs', preferences)

# Create driver using options 
driver = webdriver.Chrome(options = options)

#----------------------------------------------------------------------------------

# Visit login website 
driver.get(login_url)

# Give website time to load for slow computers 
time.sleep(3)

# Scrape the website to login using the User's info
driver.find_element_by_id('login-view-login-login').send_keys(username)
driver.find_element_by_id('login-view-login-password').send_keys(password)
driver.find_element_by_class_name('panel-body')\
.find_element_by_tag_name('button').click()

#----------------------------------------------------------------------------------

# Set scrapper to download CSV reports and set the scrapper as function 
def scrapper():
    time.sleep(3)
    
    driver.find_element_by_class_name('UnifiedExport__ddButton___2hjSR')\
    .find_element_by_tag_name('span').click()
    
    driver.find_element_by_class_name('UnifiedExport__miLink___EXLKS')\
    .find_element_by_class_name('UnifiedExport__miTitle___2P9xh').click()

# Vist each webpage and run scrapper
for each_url in reports_url:
    driver.get(each_url)
    scrapper()
```

```python
# Empty list to save DFs
dataframe_name_list = []

# Loop through CSV files in folder and create pandas DF
for filename in os.listdir(file_path):
    if filename.endswith('.csv'):
        filename = pd.read_csv(filename)
        dataframe_name_list.append(filename)
```

```python
action_df = dataframe_name_list[0]
action_df
```

```python
race_df = dataframe_name_list[1]
race_df
```

```python
reason_df = dataframe_name_list[2]
reason_df
```

```python
gender_df = dataframe_name_list[3]
result_df = dataframe_name_list[4]
```

```python
#creating new data with select columns
reason_df = reason_df[['stop_id', 'reason_for_stop', 'reason_for_stopcode', 'reason_for_stop_detail','reason_for_stop_explanation']]
gender_df = gender_df[['stop_id', 'gender']]
race_df = race_df[['stop_id', 'race']]
result_df = result_df[['stop_id', 'resultkey', 'result']]
action_df = action_df[['stop_id', 'action']]
```

```python
print(reason_df.dtypes)
```

```python
#cleaning & making sure dataframes have the same amount of rows
#combining duplicate rows where stop_id is the same
a_reason_df = reason_df.groupby('stop_id')['reason_for_stop'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
b_reason_df = reason_df.groupby('stop_id')['reason_for_stopcode'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
c_reason_df = reason_df.groupby('stop_id')['reason_for_stop_detail'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
d_reason_df = reason_df.groupby('stop_id')['reason_for_stop_explanation'].apply(lambda x: ', '.join(x.astype(str))).reset_index()

first_reason_df = pd.merge(a_reason_df, b_reason_df, on='stop_id')
second_reason_df = pd.merge(first_reason_df, c_reason_df, on='stop_id')
final_reason_df = pd.merge(second_reason_df, d_reason_df, on='stop_id')
final_reason_df
```

```python
final_race_df = race_df.groupby(['stop_id'])['race'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
final_race_df
```

```python
final_gender_df = gender_df.groupby(['stop_id'])['gender'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
final_gender_df
```

```python
e_result_df = result_df.groupby(['stop_id'])['resultkey'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
f_result_df = result_df.groupby(['stop_id'])['result'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
```

```python
final_result_df = pd.merge(e_result_df, f_result_df, on='stop_id')
final_result_df
```

```python
final_action_df = action_df.groupby(['stop_id'])['action'].apply(lambda x: ', '.join(x.astype(str))).reset_index()
final_action_df
```

```python
#connecting to local database
rds_connection_string = "postgres:postgres@localhost:5432/police_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
engine.table_names()
```

```python
final_race_df.to_sql(name='final_race_db', con=engine, if_exists='append', index=False)
```

```python
final_action_df.to_sql(name='final_action_db', con=engine, if_exists='append', index=False)
```
